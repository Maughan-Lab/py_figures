```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import py_figures_functions as pyf

from matplotlib.pyplot import rc
rc("text", usetex=True)
rc("font", **{"family":"sans-serif","sans-serif":["Helvetica"]},size="14")
rc("text.latex",preamble=r"\usepackage{sfmath}")

%matplotlib notebook
```

```python
# import data file

r, G_trunc, G_diff, G_calc = \
        np.loadtxt("C:/Users/sincl/Documents/MaughanLab/Kobi/PDF_refinements/Li3InCl6_cell3_no_mixing_refinement.txt",
        unpack=True, dtype=float, delimiter=" ", skiprows=4)
```

```python
# VARIABLES TO INPUT

# tile
title = r"Li$_3$InCl$_6$, No site mixing"
# title text position
title_pos = (1.2, 1.86)

# R-value in %
r_val = 11.45
# R-value text position
r_val_pos = (17.6, 1.35)

# experimental data
y_exp = G_trunc

# calculated fit
y_calc = G_calc
calc_color = "#FA8072"

# difference curve
diff = G_diff
# include difference curve?
addDiff = True

# axis limits
x_ax = (1, 22)
y_ax = (-1.25, 2)

# save a PNG image of graph? 
savePNG = False
# file name
fn = "cell3_PDF_no_mixing"
# save directory
save_dir = "C:/Users/sincl/Desktop/"
```

```python
# generates plot
plt.figure(figsize=(7,7))

plt.plot(r, y_exp, color="black", label="Observed", marker=".")
plt.plot(r, y_calc, color=calc_color, label="Calculated", linewidth="2")
if addDiff == True:
    plt.plot(r, diff-1, color="#BEBEBE", label="Difference")

plt.xlim(x_ax[0], x_ax[1])
plt.ylim(y_ax[0], y_ax[1])
plt.xlabel("r / " r"$\AA$")
plt.ylabel("G(r) / " r"$\AA^{-2}$")

plt.text(title_pos[0], title_pos[1], title)
plt.text(r_val_pos[0], r_val_pos[1], "R" r"$_W=$" + str(r_val) + "\%")

plt.legend()
plt.tight_layout()

if savePNG == True:
    plt.savefig(save_dir + fn + ".png", bbox_inches='tight', dpi=1000) 
```
